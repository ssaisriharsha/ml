```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from torch.nn.functional import softmax
import torch
```

```python
model_name="openai-community/gpt2"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_name)
```

```python
model=AutoModelForCausalLM.from_pretrained(model_name).to("cuda")
```

```python
tokenizer.pad_token="<PAD>"
enc_inp=tokenizer(["Today, we are going to",
           "The project we are going to do is"],
          truncation=True,
          padding=True,
          return_tensors="pt",
          padding_side="left")
input_ids=enc_inp.input_ids.to("cuda")
attn_mask=enc_inp.attention_mask
input_ids
```

```python
for _ in range(100):
  logits=model(input_ids, attention_mask=attn_mask).logits[:,-1,:]
  # display(logits.size())
  prob=softmax(logits, dim=-1)
  # display(prob)
  next_token_id=torch.multinomial(prob, num_samples=1)
  input_ids=torch.cat([input_ids, next_token_id], dim=-1)
input_ids.size()
```

```python
input_ids[0]
tokenizer.decode(input_ids[0],
                  skip_special_tokens=True)
```

```python
import torch
from torch.optim import AdamW
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
sequences=["This is a very good book", "I didn't like this movie"]
model_name="bert-base-uncased"
model=AutoModelForSequenceClassification.from_pretrained(model_name).to("cuda")
tokenizer=AutoTokenizer.from_pretrained(model_name)
enc_inps=tokenizer(sequences,
                   truncation=True,
                   padding="longest",
                   return_tensors="pt").to("cuda")
enc_inps["labels"]=torch.tensor([1, 1]).to("cuda")
display(enc_inps)
optimizer=AdamW(model.parameters())
optimizer.zero_grad()
loss=model(**enc_inps).loss
loss.backward()
optimizer.step()
```
